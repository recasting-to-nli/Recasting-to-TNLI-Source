```python
data_path = './data/fetaQA-v1_train.jsonl'
f = open(data_path,'r')
```

```python
import json,random

new_data = []
null_values = ['-','null','None','none','TBA','tba','','—']
for line in f:
    data = json.loads(line)
    data['contradictions'] = []
#     print(data)
    found = []
    all_contra = []
    for cell in data['highlighted_cell_ids']:
        row, col = cell[0], cell[1]
        text = data['table_array'][row][col]
        contra = []
        if text.lower() in data['answer'].lower(): 
            ind = data['answer'].lower().index(text.lower())
            matching_portion = data['answer'][ind:ind+len(text)]
            found.append(text)
            candidates = []
            for r in range(1,len(data['table_array'])):
                if data['table_array'][r][col] not in candidates \
                    and data['table_array'][r][col] != text \
                    and data['table_array'][r][col] not in null_values:
                    candidates.append(data['table_array'][r][col])
            if len(candidates) > 0:
                rep = random.sample(candidates, min(2,len(candidates)))
                for replacement in rep: 
                    contra.append(data['answer'].replace(matching_portion, replacement))
        if contra != [] : all_contra.append(contra)
    if len(all_contra) > 1:
        count = 1
        for contra_set in all_contra:
            if count>2: break
            sent = contra_set[0]
            del contra_set[0]
            data['contradictions'].append(sent)
            count += 1
    elif len(all_contra) > 1: data['contradictions'] = data['contradictions'] + all_contra[0]
    else : pass
    new_data.append(data)
#     print(data['contradictions'])
```

```python
entail = 0
contra = 0
no_contra = 0

index = 0
for d in new_data:
    
    entail += 1
    contra += len(d['contradictions'])
    if len(d['contradictions']) == 0: 
        no_contra += 1
#         print(index)
#         print(d)
    index += 1
    
```
