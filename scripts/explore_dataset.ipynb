```python
path = open('squall_v2/squall_v2/train_split/train_data.json','r')
tab = 'squall_v2/squall_v2/train_split/train_csv/'
```

```python
import json

data = json.load(path)
```

```python
from IPython.display import display
import pandas as pd

target = 1560
count = 0
for key,value in data.items():
    count += 1
    if count == target:
        table = pd.read_csv(tab + key)
        display(table)
        for sent,lbl in zip(value[0],value[1]):
            if lbl == 1: label = "Entail"
            elif lbl == 0: label = "Refute"
            print(label, " : ", sent)
    
```
